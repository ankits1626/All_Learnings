```python
import requests
import base64
import os
```

```python
def _prepare_payload(file_path, receipt_number, brand, date_of_receipt, brand_model, client, response_id = None):
        with open(file_path, "rb") as image_file:
            encoded_string = base64.b64encode(image_file.read()).decode("utf-8")
        payload = {
                "receipt_number": receipt_number,
                "receipt_date": date_of_receipt,  # Assuming receipt_date is a date object
                "brand": brand,
                "brand_model": brand_model,
                "encoded_receipt_file": encoded_string,
                "receipt_client": client
            }
        if response_id:
             payload['response_id'] = response_id 
        # Create a copy of the payload and remove the 'encoded_receipt_file' before printing
        printable_payload = {key: value for key, value in payload.items() if key != 'encoded_receipt_file'}
        print('<<<<< payload \n {}'.format(printable_payload))  
        return payload 
```

```python
# id,date_of_receipt,receipt_number,remarks,units,brand,category,brand_model,status,img,file_exists,file_name,file_size_MB
# 653543,6/10/24,185953,Approved,1,Ray-Ban,Optical,rb8772,1,Receipt/2024/x2cr4u2o27.jpg,Yes,x2cr4u2o27.jpg,4.433768
# data_folder = '/Users/ankit/code/innovations_docs/auto_approve/lux_dataset/updated_dataset/ds_10072024/classifier/test/hand'
data_folder = 'dryice_test/resized_folder'
receipt_number = '16907199'
brand = 'Prada'
date_of_receipt = '1/9/24' #2024-09-01
file_path = '0d7snp9uvc.jpg'
file_path = os.path.join(data_folder, file_path)
brand_model = 'SPR-SPRA12F-16K08Z-54'
client = 'Luxottica'
# response_id = '1f86ff17-a5e4-41bc-b931-ec33fa09b290'
response_id = None
```

```python
api_url = 'http://0.0.0.0:8003/receipts/validate-receipt'
headers = {'Content-Type': 'application/json'}
payload = _prepare_payload(file_path, receipt_number, brand, date_of_receipt, brand_model, client, response_id)
# Make the HTTP request
response = requests.post(api_url, json=payload, headers=headers)
print(response)
```

```python
import json
try:
    response_json = response.json()  # This will parse the JSON content from the response
    print("Response JSON:", json.dumps(response_json, indent=4))  # Pretty print the JSON response
except json.JSONDecodeError:
    print("Response content is not in JSON format:", response.text)
```
